# Getting rid of bullet points in Used name column

First check is what is the length of Used name column

```python
len(df)
```

is it the same as the number of entries having the bullet point in it?

```python
templist = df['Used name']
notokrows = 0
for item in templist:
    if "•" in str(item):
        notokrows += 1
print(notokrows)
```

so all rows has the bullet point in it.<br>
Alternative way to check this is given below.<br>str.contains() method in pandas can be used here.

```python
df['Used name'].str.contains("•")
```

We create a name column copying everything else from the Used name column except the weird symbol

```python
df['name'] = df['Used name'].str.replace("•"," ")
```

```python
df[['Height', 'Weight']] = df['Measurements'].str.split(' / ', expand=True)
```
